```python
from quick_calib.camera import get_vp_from_calib
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

root_dir = Path("../data/synth_road")
labels = np.load(root_dir / "labels.npy")
IDX = 16

# Load the first image
first_image_path = root_dir / "canny/" / f"{IDX}.png"
print(first_image_path)
image = Image.open(first_image_path)
label = labels[IDX]

# Compute vanishing points (assuming 'labels' contains the required calibration data)
# You might need to adjust this based on your specific calibration data format
vp = get_vp_from_calib(label[0], -label[1])
print(vp)

# Plot vanishing points on the image
plt.imshow(image)
plt.scatter(x=vp[0], y=vp[1], c='red', marker='x')  # Adjust color and marker style as needed

# Show image with vanishing points
plt.show()
```
